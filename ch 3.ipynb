```python
from skimage import io
img=io.imread('images/parrot.jpg')
io.imshow(img)
```

```python
#Getting Image Resolution
img.shape
```

```python
# Getting pixel values
import pandas as pd
df=pd.DataFrame(img.flatten())
df
```

```python
# Converting Color Spaces
from skimage import color, data
from pylab import *
img_hsv= color.rgb2hsv(img)
figure(0)
io.imshow(img_hsv)
figure(1)
io.imshow(color.rgb2grey(img))
```

```python
# Saving
io.imsave('images/grey-parrrot.jpg', color.rgb2gray(img))
```

```python
# Gamma Corresction
from skimage import exposure
gamma_corrected1 = exposure.adjust_gamma(img, 0.5)
gamma_corrected2 = exposure.adjust_gamma(img, 5)
figure(0)
io.imshow(gamma_corrected1)
figure(1)
io.imshow(gamma_corrected2)
```

```python
# Rotating, Shifting, and Scaling Images
from skimage.transform import rotate
img_rotate= rotate(img, 20)
io.imshow(img_rotate)
plt.axis('off')
```

```python
from skimage.transform import resize
img_resize= resize(img, (100,100))
io.imshow(img_resize)
```
